## Custom function

```python
def AddTwoNumbers(x, y):
    """
    This function is to add x and y numbers
    """
    return x + y
```

```python
AddTwoNumbers(2, 9)
```

```python
def compare(x, y):
    """
    This function is to compare 2 numbers
    """
    if x > y:
        print(f"{x} is greater than {y}")
    elif x < y:
        print(f"{x} is smaller than {y}")
    else:
        print(f"{x} is equal to {y}")
```

```python
compare(2, 2)
```

## Create custom image

```python
import matplotlib.pyplot as plt
```

```python
black = np.zeros((200, 200), dtype=np.uint8)
plt.imshow(black, cmap=plt.cm.gray)    # cmap (colormap)
plt.title("grayscale black img")
plt.xticks([]), plt.yticks([])     # x value, y value of axis
plt.show()
```

```python
white = 255 * np.ones((200, 200))
white = np.uint8(white)
# white = white.astype("uint8")
plt.imshow(white, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.title("grayscale white img")
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
random_color_img = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8)
plt.imshow(random_color_img, cmap=plt.cm.gray)
plt.title("random noise color img")
plt.xticks([]), plt.yticks([])
plt.show()
```

Exercise 1

```python
def oddOrEven(x):
    """To distinguish odd or even number"""
    if isinstance(x, int):
        if x % 2 == 0:
            if x == 0: 
                statement = " is neither even nor odd"
            else:
                statement = " is even"
        else:
            statement = " is odd"
    else:
        statement = " is not an integer"
    print(str(x) + statement)
```

```python
oddOrEven(0)
```

```python
large_n = 1e6

numpy_arr = np.arange(large_n)
list_arr = list(range(int(large_n)))

print(numpy_arr[-5:])
print(list_arr[-5:])
```

```python
%timeit res = 2*numpy_arr
```

```python
ex_black_img = np.zeros((200, 200), dtype=np.uint8)
ex_white_circle = plt.Circle((100, 100), 20, color="#fff")

fig, ax = plt.subplots()
plt.title("black img with white circular spot at the middle")
plt.xticks([]), plt.yticks([])

plt.imshow(ex_black_img, cmap=plt.cm.gray, vmin=0, vmax=255)
ax.add_patch(ex_white_circle)

plt.show()
```
